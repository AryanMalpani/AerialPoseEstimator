```python
import cv2
import numpy as np
import os

def gammaCorrection(src, filename):
    parent = "/Users/adityasomani/Desktop/3-1/CV/Destination/"
    direc = str(filename)+"/"
    path = os.path.join(parent, direc) 
    os.mkdir(path)
    gammaMatrix = [2,3,4,5,6,7,8,9,10]
    for gamma in gammaMatrix:
        invGamma = 1 / gamma

        table = [((i / 255) ** invGamma) * 255 for i in range(256)]
        table = np.array(table, np.uint8)
        file = path+str(filename)+"_"+ str(gamma)+".jpeg"
        cv2.imwrite(file, cv2.LUT(src, table))
```

```python

folder = "/Users/adityasomani/Desktop/3-1/CV/Project"


for filename in os.listdir(folder):
    temp = os.path.join(folder,filename)
    if(temp.endswith('jpeg')):
        img = cv2.imread(temp)
        gammaCorrection(img, filename)
```
